```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df_cbeskul = pd.read_excel('dataset/Clean_Dataset_Content.xlsx')
df_cbeskul.head()
```

```python
print('Banyak data:', len(df_cbeskul.Nama.unique()))
print('Jurusan:', df_cbeskul['Jurusan'].unique())
print('Jenis Kelamin :', df_cbeskul['Jenis Kelamin '].unique())
print('Alasan memilih jurusan :', df_cbeskul['Alasan memilih jurusan '].unique())
```

```python
df_cbeskul = df_cbeskul[['Nama', 'Jenis Kelamin ', 'Asal Sekolah ', 'Jurusan', 'Alasan memilih jurusan ', 'Minat_kemampuan', 'Apakah Minat dan kemampuan anda sudah sesuai dengan jurusan saat ini ']]
df_cbeskul.head()
```

```python
dropped_columns = ["Jenis Kelamin ", "Alasan memilih jurusan ", "Apakah Minat dan kemampuan anda sudah sesuai dengan jurusan saat ini ", "Asal Sekolah "]
df_cbeskul = df_cbeskul.drop(dropped_columns, axis=1)
```

```python
df_cbeskul.head()
```

```python
df_cbeskul['Minat_kemampuan'].value_counts()
```

```python
cv = CountVectorizer()
X = cv.fit_transform(df_cbeskul['Minat_kemampuan'])
```

```python
nb = MultinomialNB()
nb.fit(X, df_cbeskul['Jurusan'])
```

```python
new_data = [
    'Cenderung lebih mudah bekerja dan berpikir mengenai hal-hal yang berhubungan dengan angka'
    ]
X_new = cv.transform(new_data)
y_pred = nb.predict(X_new)
print(y_pred)
```

```python
df_cbeskul.set_index('Nama', inplace = True)
df_cbeskul.head()
```

```python
from rake_nltk import Rake
import nltk
nltk.download('stopwords')
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords

df_cbeskul['Key_words'] = ""

for index, row in df_cbeskul.iterrows():
  plot = row['Minat_kemampuan']
  r = Rake()
  r.extract_keywords_from_text(plot)

  key_words_dict_scores = r.get_word_degrees()

  row['Key_words'] = list(key_words_dict_scores.keys())

# dropping the Plot column
df_cbeskul.drop(columns = ['Minat_kemampuan'], inplace = True)
df_cbeskul.head()
```

```python
list_stopwords = set(stopwords.words('indonesian'))
def stopwords_removal(words):

  return [word for word in words if word not in list_stopwords]
  
df_cbeskul['Key_words'] = df_cbeskul['Key_words'].apply(stopwords_removal)
df_cbeskul
```

```python
df_cbeskul['bag_of_words'] = ''
columns = df_cbeskul.columns
for index, row in df_cbeskul.iterrows():
    words = ''
    for col in columns:
      if col == 'Key_words':
              words = words + ' '.join(row[col]) + ''
      else:
              words = words + row[col] + ' '
    row['bag_of_words'] = words

df_cbeskul.drop(columns = [col for col in df_cbeskul.columns if col!= 'bag_of_words'], inplace = True)

df_cbeskul.head()
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df_cbeskul['bag_of_words'])
```

```python
count.get_feature_names()
```

```python
cv_count = pd.DataFrame(count_matrix.toarray())
cv_count
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(count_matrix)
cosine_sim
```

```python
indices = pd.Series(df_cbeskul.index)

def recommendations(nama, cosine_sim = cosine_sim):

    recommended = []


    idx = indices[indices == nama].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    top_indexes = list(score_series.iloc[1:2].index)

    for i in top_indexes:
        recommended.append(list(df_cbeskul['bag_of_words'])[i])

    return recommended
```

```python
x = recommendations('Mahasiswa200')
for i in x:
  print(i.split()[0])
```
